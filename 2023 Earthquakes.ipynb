```python
counts = df['country'].value_counts()
(counts/100).plot(kind='bar', title='Earthquake Frequency by Country', figsize=(45, 25))
plt.ylabel('Earthquakes (in hundreds)')

plt.show()
```

```python
top_countries = df['country'].value_counts().head(20)
plt.figure(figsize=(30, 20))
sns.barplot(x=top_countries.index, y=top_countries)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Earthquakes')
plt.title('Top 20 Countries with Most Earthquakes')
plt.show()
```

```python
norm = Normalize(vmin=1, vmax=7)
cmap = sns.color_palette("mako_r", as_cmap=True)
colors = cmap(norm(df['mag']))
plt.figure(figsize=(30, 30))
plt.scatter(x=df['mag'], y=df['country'], s=35, c=colors)
cbar = plt.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap), orientation='vertical')
cbar.set_label('Magnitude')

plt.xlabel('Magnitude')
plt.title('Earthquake Magnitude Frequency Per Country')
plt.show()
```

```python
top_magnitude = df['mag'].value_counts().head(20)
plt.figure(figsize=(15, 10))
sns.barplot(x=top_magnitude.index, y=top_magnitude)
plt.xlabel('Magnitude')
plt.ylabel('Earthquakes')
plt.title('Magnitude with Highest Frequency')
plt.show()
```
